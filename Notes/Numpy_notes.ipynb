```
# Notes

#import numpy
import numpy as np

#create array from list
lst = [1,2,3]
arr = np.array(lst)
print(arr)

#sum
arr2 = arr+1
print(arr2)

#mult
arr3 = arr2*5
print(arr3)
```

```
#arange: Return evenly spaced values within a given interval.
#similar to range python
from0To50Every2 = np.arange(0,50,2,np.int16) #data type will be int16

#linspace: Return evenly spaced numbers over a specified interval.
from0To10 = np.linspace(0,10,50,True,False)
print(from0To10)
```

```
#eturn a new array of given shape and type, filled with zeros or ones. 
print( np.zeros((5,5)) )
print( np.ones((6,18)) )
```

```
#multi dimensional array
arr_2d = np.array([[1,2,4],[3,4,3]])

#Use shape to know how many columns and rows the arr has.
arr_2d.shape #arr_2d will return 2,3

arr_3d = np.array(  [[1,2],[2,-22],[2,1],[3,0]] )
arr_3d.shape #arr_3d will return 4,2

#size: how many elements the array contains
arr_2d.size

#max and min
arr_2d.max()
arr_3d.min()

arr_3d.max(axis=1) # will return the max of every row
arr_3d.min(axis=0) # will return the min of every column
```

```
#reshape

arr_2d = np.arange(9) # [1,2,3,4,5,6,7,8,9]
arr_2d_reshaped = arr_2d.reshape(3,3)
print(arr_2d_reshaped) # [1,2,3],[4,5,6],[6,7,8]

#slicing
arr_2d_reshaped[:2,:2] 
```
